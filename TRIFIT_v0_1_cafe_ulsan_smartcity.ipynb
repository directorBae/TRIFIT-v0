```python
def delete_newline(w):
    w = w.strip()
    w = re.sub(r"[^0-9가-힣?.!,¿]+", " ", w)
    w = w.strip()
    return w
```

```python
def preprocess(text):
    text = delete_newline(text)
    text = okt.normalize(text)
    morp = okt.morphs(text, stem=True)
    postag = okt.pos(text, stem=True)
    sentence_end = []
    for i in range(len(postag)):
        if postag[i][1] == 'Punctuation' and postag[i][0]!=',':
            sentence_end.append(i)
    return [text, postag, sentence_end, morp]
```

```python
def preprocess_wout_norm(text):
    text = delete_newline(text)
    text = okt.normalize(text)
    morp = okt.morphs(text, stem=False)
    postag = okt.pos(text, stem=False)
    sentence_end = []
    for i in range(len(postag)):
        if postag[i][1] == 'Punctuation' and postag[i][0]!=',':
            sentence_end.append(i)
    return [text, postag, sentence_end, morp]
```

```python
def count_keys(text, morp):
    morp_list = []
    for word in morp:
        morp_list.append(word[0])
    
    i=0
    gkeys_normal = []

    for gkey in gkeys:
        gkeys_normal.append([])
        for word in gkey:
            postag = okt.pos(word, stem=True)
            if postag[0][1] == 'Adjective' or postag[0][1] == 'Adverb' or postag[0][1] == 'Noun' or postag[0][1] == 'Verb':
                gkeys_normal[i].append(postag[0][0])
        i+=1
    
    intersect_list = []
    
    for i in range(6):
        intersect_list.append(list(set(morp_list) & set(gkeys_normal[i])))
    
    return intersect_list
```

```python
def keys_location(intersection, morp_wout_tag):
    keys_loc = []
    i=0
    for keys in intersection:
        keys_loc.append([])
        for word in keys:
            try:
                keys_loc[i].append(list(np.where(np.array(morp_wout_tag) == word)[0]))
            except:
                pass
        i+=1
    return keys_loc
```

```python
def count_keys_in_text(intersection, morp_wout_tag):
    i=0
    count_keys_list = [0, 0, 0, 0, 0, 0]
    for keys in intersection:
        for word in keys:
            try:
                count_keys_list[i]+=((int(morp_wout_tag.count(word)))*dict2_w.loc[word, "newavg"])
            except: pass
        i+=1
    return count_keys_list
```

```python
def normalize(lst):
    new = []
    for item in lst:
        try:
            new.append(item/sum(lst))
        except:
            return new
    return new
```

```python
def count_keys_in_each_sentence(text):
    func_return_list = preprocess(text)
    text = func_return_list[0]
    morp = func_return_list[1]
    end = func_return_list[2]
    morp_wout_tag = func_return_list[3]
    intersection = count_keys(text, morp)
    keys_location_list = keys_location(intersection, morp_wout_tag)
    keys_location_sentence = []
    count_keys_in_text_output = count_keys_in_text(intersection, morp_wout_tag)
    return count_keys_in_text_output
```

```python
def phrase_extracter(raw_key):
    part = ['Adjective', 'Adverb', 'Noun', 'Verb']
    gkeys_phrase = []
    i=0
    for gkey in raw_key:
        gkeys_phrase.append([])
        j=0
        for word in gkey:
            postag = okt.pos(word, stem=True)
            if len(postag)>1:
                gkeys_phrase[i].append([])
                for index in postag:
                    gkeys_phrase[i][j].append(index[0])
                j+=1
            else:
                pass
        i+=1
    return gkeys_phrase
```

```python
def count_phrases(text, morp_wout_tag, gkeys_phrase):
    
    phrase_totext = []
    
    for i in range(6):
        phrase_totext.append([])
    
    i=0
    for phrase_list in gkeys_phrase:
        for phrase in phrase_list:
            phrase_totext[i].append(''.join(phrase))
        i+=1
    
    count_list = [0,0,0,0,0,0]
    
    text_tocpr = ''.join( morp_wout_tag)
    
    for i in range(6):
        for item in phrase_totext[i]:
            if item in text_tocpr:
                count_list[i]+=1
    
    return count_list
```

```python
def count_phrases_in_each_sentence(text):
    func_return_list = preprocess(text)
    text = func_return_list[0]
    morp = func_return_list[1]
    end = func_return_list[2]
    morp_wout_tag = func_return_list[3]
    gkeys_phrase = phrase_extracter(gkeys)
    return normalize(count_phrases(text, morp_wout_tag, gkeys_phrase))
```

```python
def morp_to_sentence(morp_nx, sentence_end):
    sentence_list = []
    stc = ""
    index = -1
    for item in sentence_end:
        for j in range(index+1, item):
            if morp_nx[j][1] != 'Josa' and morp_nx[j][1] != 'Suffix' and j != index+1:
                stc += (" "+morp_nx[j][0])
            else:
                stc += morp_nx[j][0]
        sentence_list.append(stc)
        index=item
        stc = ""
    return sentence_list
```

```python
def count_sentences(stc1, stc2):
    from sklearn.feature_extraction.text import TfidfVectorizer
    sentences = (stc1, stc2)
    tfidf_vectorizer = TfidfVectorizer()

    tfidf_matrix = tfidf_vectorizer.fit_transform(sentences)

    from sklearn.metrics.pairwise import cosine_similarity

    cos_similar = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])
    
    from sklearn.metrics.pairwise import euclidean_distances

    tfidf_normalized = tfidf_matrix/np.sum(tfidf_matrix)

    euc_d_norm = euclidean_distances(tfidf_normalized[0:1],tfidf_normalized[1:2])

    from sklearn.metrics.pairwise import manhattan_distances
    manhattan_d = manhattan_distances(tfidf_normalized[0:1],tfidf_normalized[1:2])

    return float((cos_similar+euc_d_norm+manhattan_d)/3)
```

```python
def listsum(lst):
    output = [0,0,0,0,0,0]
    for item in lst:
        if len(item)==0:
            pass
        else:
            output = [output[i] + item[i] for i in range(len(item))]
    newoutput = [0,0,0,0,0,0]
    for j in range(len(output)):
        try:
            newoutput[j] = output[j]/len(lst)
        except:
            return newoutput
    return newoutput
```

```python
def norm(val, mean, std):
  fval = (val-mean)/std
  return fval
```

```python
def meanmax(val, min, max):
  fval = (val - min) / (max - min)
  return fval
```

```python
tag_1_norm = []
tag_2_norm = []
tag_3_norm = []
tag_4_norm = []
tag_5_norm = []
tag_6_norm = []
for item in tag_1:
  mean = np.mean(tag_1)
  std = np.std(tag_1)
  tag_1_norm.append(norm(item, mean, std))
for item in tag_2:
  mean = np.mean(tag_2)
  std = np.std(tag_2)
  tag_2_norm.append(norm(item, mean, std))
for item in tag_3:
  mean = np.mean(tag_3)
  std = np.std(tag_3)
  tag_3_norm.append(norm(item, mean, std))
for item in tag_4:
  mean = np.mean(tag_4)
  std = np.std(tag_4)
  tag_4_norm.append(norm(item, mean, std))
for item in tag_5:
  mean = np.mean(tag_5)
  std = np.std(tag_5)
  tag_5_norm.append(norm(item, mean, std))
for item in tag_6:
  mean = np.mean(tag_6)
  std = np.std(tag_6)
  tag_6_norm.append(norm(item, mean, std))
```

```python
tag_1_final = []
tag_2_final = []
tag_3_final = []
tag_4_final = []
tag_5_final = []
tag_6_final = []
for item in tag_1_norm:
  min = np.min(tag_1_norm)
  max = np.max(tag_1_norm)
  tag_1_final.append(meanmax(item, min, max))
for item in tag_2_norm:
  min = np.min(tag_2_norm)
  max = np.max(tag_2_norm)
  tag_2_final.append(meanmax(item, min, max))
for item in tag_3_norm:
  min = np.min(tag_3_norm)
  max = np.max(tag_3_norm)
  tag_3_final.append(meanmax(item, min, max))
for item in tag_4_norm:
  min = np.min(tag_4_norm)
  max = np.max(tag_4_norm)
  tag_4_final.append(meanmax(item, min, max))
for item in tag_5_norm:
  min = np.min(tag_5_norm)
  max = np.max(tag_5_norm)
  tag_5_final.append(meanmax(item, min, max))
for item in tag_6_norm:
  min = np.min(tag_6_norm)
  max = np.max(tag_6_norm)
  tag_6_final.append(meanmax(item, min, max))         
```
